```python
from sklearn.model_selection import train_test_split
```

```python
Regression.fit(x_train,y_train)
```

```python
Regression.intercept_
```

```python
Regression.predict
```

```python
#R2
R =Regression.score(x_test,y_test)
X.columns
```
